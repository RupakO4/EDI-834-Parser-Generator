```python
def parse_edi_834(file_path):
    
    with open(file_path, "r") as f:
        data = f.read()
#         print(data)
    
    segments = data.split("~")
    
    result_dict = {}
    
    for segment in segments:
        segment_data = segment.split('*')

        #key = segment_data[0].strip() #without strip
        key = segment_data[0]

        #i += 1 
        #values = [value.strip() for value in segment_data[1:]]
        
        if key == "ISA":
            data = isa_data(segment_data)
        elif key == "NM1":
#             print('nm1')
            data = NM1_MemberName(segment_data)
        elif key == "PER":
#             print('per')
            data = PER_MemberCommunicationsNumbers(segment_data)
        elif key == "N3":
#             print('n3')
            N3_MemberResidenceStreetAddress(segment_data)
        elif key == "N4":
#             print('n4')
            N4_MemberCity_State_ZIPCode(segment_data)
        elif key == "DMG":
#             print('dmg')
            DMG_MemberDemographics(segment_data)
        
        if key in result_dict:
            key = f"{key}_{segment_data[1]}"
        

        #values.insert(0, segments.index(segment))
        #values.append(segments.index(segment))
        
        #values.append(i)
#         result_dict[key].append(data)
        result_dict[key] = data
    json_data = json.dumps(result_dict , indent = 4)
        
    print(json_data)
    return json_data
```

```python
def isa_data(segment_data = None):
    isa_keys = ["AuthorizationInformationQualifier_1",
                        "AuthorizationInformation_2",
                        "SecurityInformationQualifier_3",
                        "SecurityInformation_4",
                        "SenderIDQualifier_5",
                        "InterchangeSenderID_6",
                        "ReceiverIDQualifier_7",
                        "InterchangeReceiverID_8",
                        "InterchangeDate_9",
                        "InterchangeTime_10",
                        "InterchangeControlStandardsIdentifier_11",
                        "InterchangeControlVersionNumber_12",
                        "InterchangeControlNumber_13",
                        "AcknowledgementRequested_14",
                        "UsageIndicator_15",
                        "ComponentElementSeparator_16"]
    isa_values = [value for value in segment_data[1:]] # without stirp()
    isa_dict = {k: v for k, v in zip(isa_keys, isa_values)}
    
    return isa_dict
#     json_data = json.dumps(isa_dict , indent = 4)
    
#     print(json_data)
#     print(isa_dict)
```

```python
from typing import Any, TypeVar, Type, cast


T = TypeVar("T")


def from_str(x: Any) -> str:
    assert isinstance(x, str)
    return x


def to_class(c: Type[T], x: Any) -> dict:
    assert isinstance(x, c)
    return cast(Any, x).to_dict()


class NM1:
    entity_identifier_code_01: str
    entity_type_qualifier_02: int
    response_contact_lastor_organization_name_03: str
    response_contact_first_name_04: str
    response_contact_middle_name_05: str
    identification_code_qualifier_08: int
    response_contact_identifier_09: int

    def __init__(self, entity_identifier_code_01: str, entity_type_qualifier_02: int, response_contact_lastor_organization_name_03: str, response_contact_first_name_04: str, response_contact_middle_name_05: str, identification_code_qualifier_08: int, response_contact_identifier_09: int) -> None:
        self.entity_identifier_code_01 = entity_identifier_code_01
        self.entity_type_qualifier_02 = entity_type_qualifier_02
        self.response_contact_lastor_organization_name_03 = response_contact_lastor_organization_name_03
        self.response_contact_first_name_04 = response_contact_first_name_04
        self.response_contact_middle_name_05 = response_contact_middle_name_05
        self.identification_code_qualifier_08 = identification_code_qualifier_08
        self.response_contact_identifier_09 = response_contact_identifier_09

    @staticmethod
    def from_dict(obj: Any) -> 'NM1':
        assert isinstance(obj, dict)
        entity_identifier_code_01 = from_str(obj.get("EntityIdentifierCode_01"))
        entity_type_qualifier_02 = int(from_str(obj.get("EntityTypeQualifier_02")))
        response_contact_lastor_organization_name_03 = from_str(obj.get("ResponseContactLastorOrganizationName_03"))
        response_contact_first_name_04 = from_str(obj.get("ResponseContactFirstName_04"))
        response_contact_middle_name_05 = from_str(obj.get("ResponseContactMiddleName_05"))
        identification_code_qualifier_08 = int(from_str(obj.get("IdentificationCodeQualifier_08")))
        response_contact_identifier_09 = int(from_str(obj.get("ResponseContactIdentifier_09")))
        return NM1MemberName(entity_identifier_code_01, entity_type_qualifier_02, response_contact_lastor_organization_name_03, response_contact_first_name_04, response_contact_middle_name_05, identification_code_qualifier_08, response_contact_identifier_09)

    def to_dict(self) -> dict:
        result: dict = {}
        result["EntityIdentifierCode_01"] = from_str(self.entity_identifier_code_01)
        result["EntityTypeQualifier_02"] = from_str(str(self.entity_type_qualifier_02))
        result["ResponseContactLastorOrganizationName_03"] = from_str(self.response_contact_lastor_organization_name_03)
        result["ResponseContactFirstName_04"] = from_str(self.response_contact_first_name_04)
        result["ResponseContactMiddleName_05"] = from_str(self.response_contact_middle_name_05)
        result["IdentificationCodeQualifier_08"] = from_str(str(self.identification_code_qualifier_08))
        result["ResponseContactIdentifier_09"] = from_str(str(self.response_contact_identifier_09))
        return result
```
